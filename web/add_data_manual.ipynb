```python
import pandas as pd
```

```python
path = r"C:\Users\thuyl\Downloads\lln_excel_subs_2025-2-8_6540984.xlsx"
# Thêm thông tin video (Giả sử ID và tiêu đề cố định)
video_id = "kD3-DKkiVeA"
video_title = "Why you feel stuck — and how to get motivated - Shannon Odell"

def time_to_seconds(time_str):
    """ Chuyển đổi định dạng thời gian hh:mm:ss hoặc mm:ss thành số giây """
    if "s" in time_str:
        return int(time_str.split("s")[0])
    
    parts = list(map(float, time_str.split(":")))

    if len(parts) == 3:  # hh:mm:ss
        hours, minutes, seconds = parts
    elif len(parts) == 2:  # mm:ss
        hours = 0
        minutes, seconds = parts
    else:
        raise ValueError(f"Định dạng thời gian không hợp lệ: {time_str}")

    return hours * 3600 + minutes * 60 + seconds

existing_data = pd.read_csv(r'C:\Users\thuyl\PycharmProjects\code\web_lor_tracy\web\data\youglish_data.csv')
data = pd.read_excel(path)[['Time','Subtitle']]

data["start_second"] = data["Time"].astype(str).apply(time_to_seconds)
data["end_second"] = data["start_second"].shift(-1)
data.loc[data.index[-1], "end_second"] = data["start_second"].iloc[-1] + 2


data.insert(0, "video_id", video_id)
data.insert(1, "video_title", video_title)

# Đổi tên cột Subtitle thành caption
data = data.rename(columns={"Subtitle": "caption"})

# Chỉ giữ các cột cần thiết
data = data[["video_id", "video_title", "start_second", "end_second", "caption"]]

updated_data = pd.concat([existing_data, data], ignore_index=True)
```
